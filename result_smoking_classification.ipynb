```python
import pickle
import pandas as pd
import os
import numpy as np
from plotly.offline import iplot, init_notebook_mode, plot
import plotly.graph_objs as go
```

```python
root_dir = "/Users/sasayabaku/googledrive/11010000/result"
EPOCHS = 100
```

```python
init_notebook_mode(connected=True)
```

```python
acc = pd.DataFrame(np.zeros((EPOCHS, 1)))
val_acc = pd.DataFrame(np.zeros((EPOCHS, 1)))
acc_list = pd.DataFrame(np.empty((0,0)))
val_acc_list = pd.DataFrame(np.empty((0,0)))
```

```python
f_list = os.listdir(os.path.join(root_dir))
```

```python
for i in f_list:
    with open(os.path.join(root_dir, i, 'history.log'), mode='rb') as f:
        data = pickle.load(f)
        local_acc = pd.DataFrame(data['acc'])
        local_val_acc = pd.DataFrame(data['val_acc'])
        acc += local_acc
        val_acc += local_val_acc
        acc_list = pd.concat([acc_list, local_acc], axis=1)
        val_acc_list = pd.concat([val_acc_list, local_val_acc], axis=1)
```

```python
train_y = acc_list.mean(axis=1).values
train_x = np.array(list(range(len(acc_list)))) + 1
test_y = val_acc_list.mean(axis=1).values
test_x = np.array(list(range(len(val_acc_list)))) + 1
```

```python
plot_data_acc = []
plot_val_acc = []
```

```python
trace = go.Scatter(
    x=train_x,
    y=train_y,
    name='Training',
    mode='lines+markers',
    line=dict(
        color=("#e84a5f")
    ),
    marker=dict(
        size=12,
        symbol=0
    )
)

plot_data_acc.append(trace)
```

```python
trace = go.Scatter(
    x=test_x,
    y=test_y,
    name='Test',
    mode='lines+markers',
    line=dict(
        color=("#2A363B")
    ),
    marker=dict(
        size=12,
        symbol=0
    )
)

plot_data_acc.append(trace)
```

```python
layout = go.Layout(font=dict(size=24),
                   xaxis=dict(title="epochs",
                              titlefont=dict(size=24)
                              ),
                   yaxis=dict(title="accuracy",
                              titlefont=dict(size=24)
                              )
                   )
fig = go.Figure(data=plot_data_acc, layout=layout)
```

```python
iplot(fig)
```
